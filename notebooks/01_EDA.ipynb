```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 100)
pd.set_option("display.float_format", lambda x: f"{x:.2f}")

plt.style.use("seaborn-v0_8")
sns.set_palette("husl")

# Resolve data directory relative to this notebook
NOTEBOOK_DIR = Path().resolve()
ROOT_DIR = NOTEBOOK_DIR.parent if NOTEBOOK_DIR.name == "notebooks" else NOTEBOOK_DIR
DATA_DIR = ROOT_DIR / "data" / "raw"

print(f"Root dir:  {ROOT_DIR}")
print(f"Data dir:  {DATA_DIR}")
print(f"Data exists: {DATA_DIR.exists()}")
print("Libraries loaded successfully!")
```

```python
print("Loading HC_application_train.csv...")
app_train = pd.read_csv(DATA_DIR / "HC_application_train.csv")

print(f"Shape:   {app_train.shape[0]:,} rows × {app_train.shape[1]} columns")
print(f"Memory:  {app_train.memory_usage().sum() / 1024**2:.1f} MB")
print(f"\nFirst 5 rows:")
app_train.head()
```

```python
print("=" * 50)
print("CLASS IMBALANCE ANALYSIS")
print("=" * 50)

target_counts = app_train["TARGET"].value_counts()
target_pct = app_train["TARGET"].value_counts(normalize=True) * 100

print(f"\n0 — Repaid (good):    {target_counts[0]:,} ({target_pct[0]:.1f}%)")
print(f"1 — Defaulted (bad):  {target_counts[1]:,} ({target_pct[1]:.1f}%)")
print(f"\nImbalance ratio: {target_counts[0]/target_counts[1]:.1f}:1")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Bar chart
axes[0].bar(["Repaid (0)", "Defaulted (1)"],
            target_counts.values,
            color=["#2ecc71", "#e74c3c"],
            edgecolor="white", linewidth=1.5)
axes[0].set_title("Loan Default Distribution", fontsize=14, fontweight="bold")
axes[0].set_ylabel("Count")
for i, v in enumerate(target_counts.values):
    axes[0].text(i, v + 1000, f"{v:,}\n({target_pct[i]:.1f}%)",
                 ha="center", fontsize=11)

# Pie chart
axes[1].pie(target_counts.values,
            labels=["Repaid", "Defaulted"],
            colors=["#2ecc71", "#e74c3c"],
            autopct="%1.1f%%",
            startangle=90,
            wedgeprops={"edgecolor": "white", "linewidth": 2})
axes[1].set_title("Default Rate", fontsize=14, fontweight="bold")

plt.suptitle("Target Variable — Class Imbalance", fontsize=16, fontweight="bold", y=1.02)
plt.tight_layout()

# Ensure save directory exists
SAVE_DIR = ROOT_DIR / "data" / "processed"
SAVE_DIR.mkdir(parents=True, exist_ok=True)
plt.savefig(SAVE_DIR / "01_class_imbalance.png", dpi=150, bbox_inches="tight")
plt.show()
print("\nPlot saved!")
```

```python
print("=" * 50)
print("MISSING VALUES ANALYSIS")
print("=" * 50)

# Calculate missing values
missing = app_train.isnull().sum()
missing_pct = (app_train.isnull().sum() / len(app_train)) * 100

missing_df = pd.DataFrame({
    "missing_count": missing,
    "missing_pct": missing_pct
}).sort_values("missing_pct", ascending=False)

# Only show columns with missing values
missing_df = missing_df[missing_df["missing_count"] > 0]

print(f"\nTotal columns: {app_train.shape[1]}")
print(f"Columns with missing values: {len(missing_df)}")
print(f"Columns with >50% missing: {len(missing_df[missing_df['missing_pct'] > 50])}")
print(f"Columns with >80% missing: {len(missing_df[missing_df['missing_pct'] > 80])}")

# Plot top 30 columns with most missing values
top_missing = missing_df.head(30)

plt.figure(figsize=(12, 8))
colors = ["#e74c3c" if x > 50 else "#f39c12" if x > 20 else "#3498db"
          for x in top_missing["missing_pct"]]
bars = plt.barh(range(len(top_missing)), top_missing["missing_pct"],
                color=colors, edgecolor="white")
plt.yticks(range(len(top_missing)), top_missing.index, fontsize=9) # type: ignore
plt.xlabel("Missing %")
plt.title("Top 30 Columns by Missing Value %", fontsize=14, fontweight="bold")

# Add legend
from matplotlib.patches import Patch  # noqa: E402
legend_elements = [
    Patch(facecolor="#e74c3c", label=">50% missing"),
    Patch(facecolor="#f39c12", label="20-50% missing"),
    Patch(facecolor="#3498db", label="<20% missing")
]
plt.legend(handles=legend_elements, loc="lower right")
plt.tight_layout()
plt.savefig(SAVE_DIR / "02_missing_values.png", dpi=150, bbox_inches="tight")
plt.show()
print("\nTop 10 columns with most missing values:")
print(missing_df.head(10).to_string())
```

```python
print("=" * 50)
print("DATASET OVERVIEW")
print("=" * 50)

# Data types breakdown
dtype_counts = app_train.dtypes.value_counts()
print(f"\nData types:")
for dtype, count in dtype_counts.items():
    print(f"  {str(dtype):<15} {count} columns")

# Numerical vs Categorical
numerical_cols = app_train.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = app_train.select_dtypes(include=["object"]).columns.tolist()

print(f"\nNumerical columns:   {len(numerical_cols)}")
print(f"Categorical columns: {len(categorical_cols)}")

print(f"\nCategorical columns list:")
for col in categorical_cols:
    n_unique = app_train[col].nunique()
    print(f"  {col:<40} {n_unique} unique values")

# Basic stats on key financial columns
print("\n" + "=" * 50)
print("KEY FINANCIAL COLUMNS — BASIC STATS")
print("=" * 50)

key_cols = ["AMT_INCOME_TOTAL", "AMT_CREDIT", "AMT_ANNUITY", "AMT_GOODS_PRICE"]
print(app_train[key_cols].describe().round(2).to_string())
```

```python
print("=" * 50)
print("KEY FINANCIAL DISTRIBUTIONS")
print("=" * 50)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

cols = {
    "AMT_INCOME_TOTAL": "Annual Income",
    "AMT_CREDIT": "Credit Amount",
    "AMT_ANNUITY": "Annuity Amount",
    "AMT_GOODS_PRICE": "Goods Price"
}

for ax, (col, title) in zip(axes.flatten(), cols.items()):
    data = app_train[col].dropna()
    # Cap at 99th percentile to remove extreme outliers for visualization
    cap = data.quantile(0.99)
    data_capped = data[data <= cap]

    # Plot defaulters vs non-defaulters
    defaulters = app_train.loc[app_train["TARGET"] == 1, col].dropna()
    defaulters = defaulters[defaulters <= cap]
    non_defaulters = app_train.loc[app_train["TARGET"] == 0, col].dropna()
    non_defaulters = non_defaulters[non_defaulters <= cap]

    ax.hist(non_defaulters, bins=50, alpha=0.6,
            color="#2ecc71", label="Repaid", density=True)
    ax.hist(defaulters, bins=50, alpha=0.6,
            color="#e74c3c", label="Defaulted", density=True)
    ax.set_title(title, fontsize=12, fontweight="bold")
    ax.set_xlabel("Amount (₹)")
    ax.set_ylabel("Density")
    ax.legend()
    ax.xaxis.set_major_formatter(
        plt.FuncFormatter(lambda x, p: f"{x/1000:.0f}K"))

plt.suptitle("Financial Feature Distributions — Default vs Repaid",
             fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig(SAVE_DIR / "03_financial_distributions.png", dpi=150, bbox_inches="tight")
plt.show()
print("Plot saved!")
```

```python
print("=" * 50)
print("AGE & EMPLOYMENT ANALYSIS")
print("=" * 50)

# DAYS_BIRTH is negative (days before application)
# Convert to age in years
app_train["AGE_YEARS"] = abs(app_train["DAYS_BIRTH"]) / 365
app_train["EMPLOYMENT_YEARS"] = abs(
    app_train["DAYS_EMPLOYED"].replace(365243, np.nan)) / 365

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Age distribution
defaulters_age = app_train.loc[app_train["TARGET"]==1, "AGE_YEARS"]
non_defaulters_age = app_train.loc[app_train["TARGET"]==0, "AGE_YEARS"]

axes[0].hist(non_defaulters_age, bins=40, alpha=0.6,
             color="#2ecc71", label="Repaid", density=True)
axes[0].hist(defaulters_age, bins=40, alpha=0.6,
             color="#e74c3c", label="Defaulted", density=True)
axes[0].set_title("Age Distribution", fontsize=13, fontweight="bold")
axes[0].set_xlabel("Age (years)")
axes[0].set_ylabel("Density")
axes[0].legend()

# Age bin default rate
app_train["AGE_BIN"] = pd.cut(app_train["AGE_YEARS"],
                               bins=[20, 25, 30, 35, 40, 45, 50, 55, 60, 70])
age_default = app_train.groupby("AGE_BIN", observed=True)["TARGET"].mean() * 100

axes[1].bar(range(len(age_default)), age_default.values,
            color=["#e74c3c" if x > 10 else "#f39c12" if x > 8
                   else "#2ecc71" for x in age_default.values])
axes[1].set_xticks(range(len(age_default)))
axes[1].set_xticklabels([str(x) for x in age_default.index],
                         rotation=45, ha="right", fontsize=9)
axes[1].set_title("Default Rate by Age Group", fontsize=13, fontweight="bold")
axes[1].set_ylabel("Default Rate (%)")
axes[1].axhline(y=app_train["TARGET"].mean()*100,
                color="navy", linestyle="--", label="Overall avg")
axes[1].legend()

plt.suptitle("Age & Employment Analysis", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig(SAVE_DIR / "04_age_employment.png", dpi=150, bbox_inches="tight")
plt.show()

print("\nDefault rate by age group:")
print(age_default.round(2).to_string())
print(f"\nOverall default rate: {app_train['TARGET'].mean()*100:.2f}%")
```

```python
print("=" * 50)
print("CORRELATION WITH TARGET")
print("=" * 50)

# Calculate correlations with TARGET
correlations = app_train.select_dtypes(include=[np.number]).corr()["TARGET"]
correlations = correlations.drop("TARGET").sort_values()

# Top 15 positive and negative correlations
top_positive = correlations.tail(15)
top_negative = correlations.head(15)
top_corr = pd.concat([top_negative, top_positive])

fig, ax = plt.subplots(figsize=(10, 10))

colors = ["#e74c3c" if x > 0 else "#2ecc71" for x in top_corr.values]
bars = ax.barh(range(len(top_corr)), top_corr.values, color=colors)
ax.set_yticks(range(len(top_corr)))
ax.set_yticklabels(top_corr.index, fontsize=9)
ax.axvline(x=0, color="black", linewidth=0.8)
ax.set_xlabel("Correlation with TARGET (default)")
ax.set_title("Top 15 Positive & Negative Correlations with Default",
             fontsize=13, fontweight="bold")

# Add value labels
for i, (bar, val) in enumerate(zip(bars, top_corr.values)):
    ax.text(val + (0.001 if val > 0 else -0.001), i,
            f"{val:.3f}",
            va="center",
            ha="left" if val > 0 else "right",
            fontsize=8)

plt.tight_layout()
plt.savefig(SAVE_DIR / "05_correlations.png", dpi=150, bbox_inches="tight")
plt.show()

print("\nTop 10 features positively correlated with default:")
print(correlations.tail(10).round(4).to_string())
print("\nTop 10 features negatively correlated with default:")
print(correlations.head(10).round(4).to_string())
```

```python
print("=" * 50)
print("CATEGORICAL FEATURES ANALYSIS")
print("=" * 50)

cat_cols = [
    "NAME_CONTRACT_TYPE",
    "CODE_GENDER",
    "NAME_INCOME_TYPE",
    "NAME_EDUCATION_TYPE",
    "NAME_FAMILY_STATUS",
    "NAME_HOUSING_TYPE"
]

fig, axes = plt.subplots(2, 3, figsize=(16, 10))

for ax, col in zip(axes.flatten(), cat_cols):
    default_rate = app_train.groupby(col)["TARGET"].mean() * 100
    default_rate = default_rate.sort_values(ascending=True)

    colors = ["#e74c3c" if x > app_train["TARGET"].mean() * 100
              else "#2ecc71" for x in default_rate.values]

    bars = ax.barh(range(len(default_rate)),
                   default_rate.values, color=colors)
    ax.set_yticks(range(len(default_rate)))
    ax.set_yticklabels(default_rate.index, fontsize=8)
    ax.axvline(x=app_train["TARGET"].mean() * 100,
               color="navy", linestyle="--", linewidth=1, label="Avg")
    ax.set_title(col.replace("NAME_", "").replace("_", " ").title(),
                 fontsize=10, fontweight="bold")
    ax.set_xlabel("Default Rate (%)")
    ax.legend(fontsize=7)

    for i, val in enumerate(default_rate.values):
        ax.text(val + 0.1, i, f"{val:.1f}%", va="center", fontsize=7)

plt.suptitle("Default Rate by Categorical Features",
             fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig(SAVE_DIR / "06_categorical_analysis.png", dpi=150, bbox_inches="tight")
plt.show()

print("\nDefault rate by Contract Type:")
print(app_train.groupby("NAME_CONTRACT_TYPE")["TARGET"].mean().mul(100).round(2))
print("\nDefault rate by Gender:")
print(app_train.groupby("CODE_GENDER")["TARGET"].mean().mul(100).round(2))
print("\nDefault rate by Education:")
print(app_train.groupby("NAME_EDUCATION_TYPE")["TARGET"].mean().mul(100).round(2))
```
